```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math

df = pd.read_csv('GlobalLandTemperaturesByMajorCity.csv')
df
```

```python
dt = df['dt']
country = df['Country']
city = df['City']


AverageTemperature = df['AverageTemperature']
year = []
for each in dt:
#     print(each)
    y = int(each.split('-')[0])
    year.append(y)
#     print(type(each))
# print(year)
df['City']
len(city)
city.loc('Los Angeles')
la = city[131846:133823]
#print (la)
la_avg = AverageTemperature[131846:133823]
#print (la_avg)
la_year = year[131846:133823]

xmean, ymean = zip(*sorted((xVal, np.mean([yVal for a, yVal in zip(la_year, la_avg) if xVal==a])) for xVal in set(la_year)))
# print(xmean)

# plt.plot(year, AverageTemperature, 'bo')
plt.hist( la_avg, bins=10, color='red')
plt.title('Frequency of Los Angeles Temperature 1849-2013')
plt.xlabel('Temperature in C')
plt.ylabel('Months')
plt.show()

#This histogram shows the distrubition in average temperature on a monthly basis for the time stated in Los Angeles.
#Average temperatures vary on a monthly basis anywhere from 3 degrees to 28 degrees.


```

```python
xmean, ymean = zip(*sorted((xVal, np.mean([yVal for a, yVal in zip(year, AverageTemperature) if xVal==a])) for xVal in set(year)))
# print(xmean)
xmean = [ each for each in xmean if each > 1849 ]
n =len(xmean)
# print(n)
ymean = ymean[-n:]

plt.scatter(x= xmean,y =ymean)
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.title(' World Average Temperature')
plt.show()

#This scatterplot shows the World Average Temperature from 1880-2013.
#The trend appears that the earth is getting warmer as the years progress.
#The upwards trend seems to be increasing more rapidly in past few years.
```

```python
avg_uncert= df['AverageTemperatureUncertainty']
#print (avg_uncert)
xmean, ymean = zip(*sorted((xVal, np.mean([yVal for a, yVal in zip(avg_uncert, year ) if xVal==a])) for xVal in set(avg_uncert)))
#print(xmean)
xmean = [ each for each in xmean if each > 1849 ]
n =len(xmean)
#print(n)
ymean = ymean[-n:]
#print(ymean)
```

```python
plt.plot(xmean,ymean,color = 'purple')
plt.ylim([.3,1.0])
plt.ylabel('Uncertainty')
plt.xlabel ('Year')
plt.title('Temperature Uncertainty 1849:2013')
plt.show()

#This line graph shows the level of "uncertainty" in the temperature numbers over the time frame given.
#The level of uncertainty declines on a relatively steady basis.
#There are two noticable upticks in uncertainty including one currently.
```
